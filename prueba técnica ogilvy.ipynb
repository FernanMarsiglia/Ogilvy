También podemos realizar esta limpieza desde otra perspectiva. Donde filtramos por autor del tuit o retuit y solo tenemos en cuenta una vez a dicho autor en toda la base de datos. Esto desde una perspectiva de negocio es mas viable debido a que con el numero total de personas que estan mencionando nuestra marca y la cantidad de seguidores que poseen, ya sea con #, @ o simplemente hablando sobre ella, podemos calcular el alcance potencial.

```python
df3 = dfcopy.drop_duplicates(subset=["Author"], keep=False)
```

```python
df3.head(100)
```

```python
usuarios_mencionando = df3.shape[0]
print(usuarios_mencionando)
```

Por lo tanto tenemos que hay 1544 usuarios mencionando o retuiteando a nuestra marca en sus tuits. 

Para calcular el alcance potencial, tenemos en cuenta que se suma el número de usuarios que están mencionando tu marca o que retuitean una publicación de tu perfil + el número de seguidores que tienen estos usuarios.

```python
seguidores_totales = df3["Twitter Followers"].sum()
print(seguidores_totales)
```

```python
Alcance_pot = usuarios_mencionando + seguidores_totales
print(Alcance_pot)
```

Nuestro alcance potencial son 11.540.112 usuarios.

Ahora tenemos en cuenta, que para calcular la cantidad total de interacciones que hay en nuestra base de datos, hay que sumar los clics en cualquier lugar del tuit, ya sea retuits, respuestas, seguimientos, Me gusta, enlaces, cards, hashtags, contenido multimedia insertado, nombre de usuario, foto de perfil o ampliación del tuit. 
Pero en esta base de datos en particular vamos a tener en cuenta solo las columnas: 
- "Twitter Reply Count"(cantidad de veces que respondieron un tweet o retuit)	
- "Twitter Reply to" (Indica que el tuit o retuit en cuestion fue una respuesta a otro tuit)
- "Twitter Retweet of" (Indica que el tuit en cuestión es un retuit de otro)
- "Twitter Retweets" (Indica el numero de retuits que se le hicieron al tuit en cuestión)
En este punto tenemos en cuenta la base de datos total, ya que las interacciones totales es la suma de las acciones anteriormente realizadas en cada uno de los tuits.

```python
dfcopy["Twitter Reply to"].count()
```

```python
TwReplyCount = dfcopy["Twitter Reply Count"].sum()
TwReplyto = dfcopy["Twitter Reply to"].count()
TwRtof = dfcopy["Twitter Retweet of"].count()
TwRt = dfcopy["Twitter Retweets"].sum()
Interacciones = TwReplyCount + TwReplyto + TwRtof + TwRt
print(Interacciones)
print(TwReplyCount, TwReplyto, TwRtof, TwRt)
```

Se tiene un total de 10756 interacciones en la base de datos teniendo en cuenta tuits y retuits.

Ahora para calcular el día donde se recibieron mas menciones debemos separar la columna date en un formato de fecha día y otro donde este la hora.

```python
dfcopy["Dates"] = pd.to_datetime(dfcopy["Date"]).dt.date
dfcopy["Time"] = pd.to_datetime(dfcopy["Date"]).dt.time
```

```python
dfcopy.head()
```

```python
dfcopy.groupby('Dates')['Dates'].count().sort_values(ascending=False)
```

Se evidencia que el día donde hubo mas actividad fue el 30 de octubre de 2020 con 502 menciones. Evidenciando alta actividad a finales del mes de octubre y principios de noviembre.
